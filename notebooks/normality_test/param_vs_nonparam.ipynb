### These notebooks is to compare application of parametrical versus non-parametrical methods on data of varios type and size for group comparision

1. Various distributions: normal distribution and not-normal distribution (exponentional, poisson etc.)
2. Various size of data (10, 100, 1000, 5000, 20000 samples)
3. Sensitivity to outliers


The goal is to benchmark tests to see if tests provide the expected results

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats


distributions = {}

for _ in range(100):
    norm_dist1 = np.random.normal(0, 1, 10)
    norm_dist2 = np.random.normal(0, 1, 10)
    dist_10samples =[stats.f_oneway(norm_dist1, norm_dist2)[1], stats.kruskal(norm_dist1, norm_dist2)[1]]

    norm_dist3 = np.random.normal(0, 1, 100)
    norm_dist4 = np.random.normal(0, 1, 100)
    dist_100samples =[stats.f_oneway(norm_dist3, norm_dist4)[1], stats.kruskal(norm_dist3, norm_dist4)[1]]

    distributions["10 sample"] = dist_10samples
    distributions["100 samples"] = dist_100samples

distributions
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set the parameters for the Poisson distribution
mu = 5

# Generate the data for the Poisson distribution
data = np.random.poisson(mu, 1000)

# Calculate the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Create a histogram of the data
plt.hist(data, bins=range(0, 15), density=True, alpha=0.5, label='Data')

# Plot the probability mass function (PMF) of the Poisson distribution
x = range(0, 15)
pmf = poisson.pmf(x, mu)
plt.plot(x, pmf, 'ro', label='Poisson PMF')

# Add a title and labels to the plot
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for normal distribution
mu = 50
sigma = 10
data = np.random.normal(mu, sigma, 1000)

# Add some outliers
outliers = np.random.normal(20, 5, 20)
data = np.concatenate([data, outliers])

# Plot the data
fig, ax = plt.subplots()
ax.hist(data, bins=50, density=True)

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Normal Distribution with Outliers')

plt.show()

```

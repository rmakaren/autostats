```python
import numpy as np
import matplotlib.pyplot as plt

# Generate log-normal data
mu = 1.0  # Mean of the underlying normal distribution
sigma = 0.5  # Standard deviation of the underlying normal distribution
data = np.random.lognormal(mu, sigma, 1000)

# Visualize the histogram of the log-normal data
fig, ax = plt.subplots()
ax.hist(data, bins=30)
ax.set_title('Log-normal distribution')
ax.set_xlabel('Data')
ax.set_ylabel('Frequency')
plt.show()

# Apply log-transformation to the data
log_data = np.log(data)

# Visualize the histogram of the log-transformed data
fig, ax = plt.subplots()
ax.hist(log_data, bins=30)
ax.set_title('Log-transformed data')
ax.set_xlabel('Log(data)')
ax.set_ylabel('Frequency')
plt.show()
```
